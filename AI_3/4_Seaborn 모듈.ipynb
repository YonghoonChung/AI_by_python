# **2. seaborn을 사용하는 이유**

* matplotlib으로 대부분의 시각화는 가능하지만 seaborn에서만 제공되는 통계기반 plot가 존재
* 아름다운 스타일링(스타일링을 신경쓰지 않아도 default 컬러가 예쁘게 조합)
* pandas 데이터 프레임과 호환성이 좋음

```python
tips = sns.load_dataset('tips')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('violin plot')
plt.show()
```

```python
sns.countplot(tips['day'])
plt.title('countplot')
plt.show()
```

```python
sns.relplot(x='tip', y ='total_bill', data=tips)
plt.title('relplot')
plt.show()
```

```python
sns.lmplot(x='tip', y ='total_bill', data=tips)
plt.title('implot')
plt.show()
```

```python
# annot:각 cell의 값 표시 여부
# linewidths: cell사이의 선
sns.heatmap(tips.corr(), annot=True, linewidths=1)
plt.title('heatmap')
plt.show()
```

```python
sns.barplot(x = 'day', y= 'total_bill',data=tips, palette = 'colorblind')
plt.show()
```

[colorblind색상에 관하여](https://seaborn.pydata.org/tutorial/color_palettes.html)


```python
# https://chrisalbon.com/python/data_visualization/seaborn_color_palettes/

sns.palplot(sns.light_palette((210,90,60),input='husl'))
sns.palplot(sns.dark_palette('muted purple',input='xkcd'))
```

```python
sns.barplot(x = 'tip', y = 'total_bill', data=tips, palette = 'bright')
```

[palette에 대해서](https://chrisalbon.com/python/data_visualization/seaborn_color_palettes/)

```python
sns.catplot(x='sex', y='total_bill', hue='smoker', data=tips, kind='bar')
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill', 
            hue='smoker', 
            col='time',
            data=tips, 
            kind='bar')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 1000
```

```python
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
sns.scatterplot(x,y,size=area, 
                sizes=(area.min(),area.max()), 
                hue = area, 
                palette='gist_gray')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(131)
sns.scatterplot(x, y, size=area, sizes=(area.min(), area.max()), color='blue', alpha=0.1)
plt.title('alpha=0.1')
plt.legend(loc= 'lower left')
plt.subplot(132)
plt.title('alpha=0.5')
sns.scatterplot(x, y, size=area, sizes=(area.min(), area.max()), color='red', alpha=0.5)

plt.subplot(133)
plt.title('alpha=1.0')
sns.scatterplot(x, y, size=area, sizes=(area.min(), area.max()), color='green', alpha=0.9)

plt.show()
```

# **Titanic 예제 (21.12.29)**

```python
# 21.12.29
titanic = sns.load_dataset('titanic')
titanic
```

```python
# 배경을 darkgrid로 설정
sns.set(style= 'darkgrid')
```

```python
sns.countplot(x = 'class', hue='who', data= titanic)
plt.show()
```

```python
sns.countplot(y= 'class', hue = 'who', data=titanic)
plt.show()
```

```python
sns.countplot(x='class', hue = 'who' , palette='copper', data=titanic)
plt.show()
```
