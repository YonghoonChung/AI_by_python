```python
# 문제
# 1행 3열짜리 그래프 3개 작성
# 1행 1열 (100~200), 1행 2열 (200~300), 1행 3열(300~400)

data1 = np.arange(100,201)
data2 = np.arange(200,301)
data3 = np.arange(300,401)

plt.plot(data1)
plt.plot(data2)
plt.plot(data3)

plt.figure()

plt.subplot(131)
plt.plot(data1)
plt.subplot(132)
plt.plot(data2)
plt.subplot(133)
plt.plot(data3)
```

```python
# plt.subplots(행의 갯수, 열의 갯수)
data = np.arange(1, 51)

fig, axes = plt.subplots(2, 3)

axes[0, 0].plot(data)
axes[0, 1].plot(data * data)
axes[0, 2].plot(data ** 3)
axes[1, 0].plot(data % 10)
axes[1, 1].plot(-data)
axes[1, 2].plot(data // 20)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.rand(50)
```

```python
plt.figure(figsize = (12,6))

# cmap에 컬러를 설정하면 모든 그래프에 색상을 같게 설정
# alpha값은 투명도를 나타내며 0~1 사이의 값을 설정할 수 있음

plt.subplot(131)
plt.scatter(x,y,s=area, cmap = 'blue', alpha=0.1)
plt.title('alpha = 0.1')

plt.subplot(132)
plt.scatter(x,y,s=area, cmap = 'blue', alpha=0.5)
plt.title('alpha = 0.5')

plt.subplot(133)
plt.scatter(x,y,s=area, cmap = 'blue', alpha=1.0)
plt.title('alpha = 1.0')

plt.show()
```

```python
# barh 함수에서는 xticks로 설정했던 부분을 yticks로 변경
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.figure(figsize=(6, 3))
# plt.bar(x, y)
plt.barh(x, y, align='center', alpha=0.7, color='green')  # align='edge'
plt.yticks(x)
plt.xlabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```python
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

width = 0.35

fig, axes = plt.subplots()

axes.bar(x - width/2, y_1, width, align='center', alpha=0.5)
axes.bar(x + width/2, y_2, width, align='center', alpha=0.8)

plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.legend(['김사과', '반하나'])

plt.show()
```

```python
x = np.arange(0,10,0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.figure(figsize = (10,8))
plt.plot(x,y_1, label = '1+sin', alpha = 0.3)
plt.plot(x,y_2, label = '1+cos', alpha = 0.7)

plt.xlabel('x value' ,fontsize = 15)
plt.ylabel('y value' ,fontsize = 15)
plt.title('sin&cos graph' ,fontsize = 20)

plt.grid()
plt.legend()
plt.show()
```

```python
#Pie Chart

labels = ['삼성', '카카오', '로블룩스', '애플', '메타', 'AT&T']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
explode = (0.3, 0, 0, 0, 0, 0)
patches, texts, autotexts = plt.pie(
    sizes,
    explode = explode,
    labels = labels,
    autopct = '%1.1f%%',
    shadow = True,
    startangle=90
)

plt.title('성공의 지름길', fontsize=20)

# label 텍스트에 대한 스타일 적용
for t in texts:
  t.set_fontsize(12)
  t.set_color('gray')

# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts:
  t.set_color('white')
  t.set_fontsize(17)

plt.show()
```

```python
#imshow
#이미지(image) 데이터와 유사하게 행과 열을 가진 2차원의 데이터를 시각화

from sklearn.datasets import load_digits

digits = load_digits() # 0 ~16값을 가지는 array로 이루어져 있음, 8 x 8 배열
X = digits.images[:10]
X[0] # 1개의 array는 8*8 배열안에 표현, 숫자는 0~9로 표ㅎㄴ
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 5, sharex=True, figsize=(12,6), sharey = True)

for i in range(10):
  axes[i//5][i%5].imshow(X[i],cmap='Blues')
  axes[i//5][i%5].set_title(str(i), fontsize = 20)

plt.tight_layout()
plt.show()
```
