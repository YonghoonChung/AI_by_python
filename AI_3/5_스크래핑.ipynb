### **2-1. BasicEnglishSpeaking.com**

````python
import requests
from bs4 import BeautifulSoup #beautifulsoup 라이브러리에서 스크레이핑 할수있게해준다
````

````python
site='https://basicenglishspeaking.com/daily-english-conversation-topics/'
request = requests.get(site)
# print(request) #<Response [200]>
html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
````

````python
subject = []
divs = soup.findAll('div',{'class':'thrv-columns'})
print(divs)
````

````python
for div in divs:
  links = div.findAll('a')
  # print(links)
  for link in links:
    # print(link.text)
    subject.append(link.text)
````

````python
print('총', len(subject), '개의 대화를 찾았습니다')
for i in range(len(subject)):
  print('{0:2d}. {1:s}'.format(i+1, subject[i]))
````

### **2-2. 벅스뮤직 차트**

````python
request = requests.get('https://music.bugs.co.kr/chart')
html = request.text
soup = BeautifulSoup(html, 'html.parser')

titles = soup.findAll('p', {'class' : 'title'})
# print(titles)
artists = soup.findAll('p',{'class' : 'artist'})
# print(artists)
````

````python
for i in range(len(titles)):
  title = titles[i].text.strip()
  # print(title)
  artist = artists[i].text.strip().split('\n')[0]
  # print(artist)
  data = '{0:3d}위 {1} - {2}'.format(i+1, artist, title)
  print(data)
````

### **2-4 네이트판 베스트 댓글, 일반 댓글**

````python
from urllib.request import urlopen
````

````python
url = 'https://pann.nate.com/talk/364288844'
soup = BeautifulSoup(urlopen(url).read(), 'html.parser')
# print(soup)
comment_best = soup.find('div', class_='cmt_best')
comment_best = comment_best.find_all('dd', class_='usertxt')
print('👍 베스트 댓글')
for comment in comment_best:
  print(comment.get_text().strip())
  print('*'*50)

print('\n👌일반 댓글')
comments = soup.find('div',class_='cmt_list')
comments = comments.find_all('dd', class_='usertxt')
for comment in comments:
  print(comment.get_text().strip())
  print('*'*50)
````

````python
# https://news.v.daum.net/v/20220103175157964
# https://news.v.daum.net/v/20220103184616242

# 내꺼
def get_daum_news_title(news_id):
  import requests
  from bs4 import BeautifulSoup

  link = 'https://news.v.daum.net/v/'+news_id

  request = requests.get(link)
  html = request.text
  soup = BeautifulSoup(html, 'html.parser')

  title = soup.find('h3', class_='tit_view')
  print(title.get_text())

# 강사님
def get_daum_news_title(news_id):
  url = 'https://news.v.daum.net/v/{}'.format(news_id)
  request = requests.get(url)
  soup = BeautifulSoup(request.text)
  title = soup.select_one('h3.tit_view')
  if title:
    return title.get_text()
  return ""
````

````python
import datetime
import time

dt_now  = datetime.datetime.now()
curdate = dt_now.date().strftime('%Y%m%d')
# url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd='+curdate+'&hh=19&rtm=N&pg='
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20211225&hh=19&rtm=N&pg='
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
genie = []
for i in range(1, 5):
  site = '{}{}'.format(url, i)
  request = requests.get(site, headers = header)
  time.sleep(1)
  html = request.text
  soup = BeautifulSoup(html, 'html.parser')
  # print(site)
  titles = soup.findAll('a', {'class':'title ellipsis'})
  artists = soup.findAll('a', {'class' : 'artist ellipsis'})
  # print(len(titles))
  for j in range(len(titles)):
    genie.append((titles[j].text.strip().split('\n')[0],artists[j].text))

print(genie)
print(len(genie))
````

````python
import requests
link = 'https://www.nate.com/'
request = requests.get(link)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
````

````python
lists = []
issues = soup.find_all('a', class_='ik')
````

````python
for issue in issues:
    print(issue.text.split('\n')[1], end=',')
    upDown = issue.text.split('\n')[2]
    # print(upDown)
    if upDown[:2] == '상승':
        print('↑' + upDown[2:])
    elif upDown[:2] == '하락':
        print('↓' + upDown[2:])     
    elif upDown[:2] == '동일':
        print('-')
    else:
        print(upDown)
````
