### **2-3. range(numpy와는 상관없는 python문법)**
range는 범위를 지정해주는 함수
보통 for in의 반복문에서 많이 사용

```python
arr = np.arange(1, 11)
for i in arr:
  print(i)
```

```python
for i in range(1, 11, 2):
  print(i)
```

### **2-5. 행렬 - 덧셈**

```python
import numpy as np
```

```python
# 덧셈
a = np.array([[1, 2, 3],
             [2, 3, 4]])

b = np.array([[3, 4, 5],
              [1, 2, 3]])
```

```python
a + b
```

```python
a = np.array([[1, 2, 3],
             [2, 3, 4]])

b = np.array([[1, 2],
              [3, 4],
              [5, 6]])
```

```python
a + b # 다른 shape
```

```python
# sum
a = np.array([[1, 2, 3],
              [2, 3, 4]])
```

```python
np.sum(a, axis=0) #열 씩 계산
```

```python
np.sum(a, axis=1) # 행 계산
```

### **2-7. 행렬의 곱셈**

```python
a = np.array([[1, 2, 3],
              [2, 3, 4]])

b = np.array([[3, 4, 5],
              [1, 2, 3]])
```

```python
print(a.shape)
print(b.shape)
```

```python
a * b
```

> 단순 곱셈과 dot product는 다름

```python
a = np.array([[1, 2, 3],
              [1, 2, 3],
              [2, 3, 4]])

b = np.array([[1, 2],
              [3, 4],
              [5, 6]])
```

```python
a.shape, b.shape
```

```python
np.dot(a,b)
```
