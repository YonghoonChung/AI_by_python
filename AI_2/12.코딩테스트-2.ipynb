# **1. 나이순 정렬**
* https://www.acmicpc.net/problem/10814


```python
n = int(input())

array = []

for _ in range(n):
  data = input().split(' ')
  array.append((int(data[0]), data[1]))

# print(array)
# sorted(리스트, key=lambda식) : key에 제공된 값을 기준으로 정렬
array = sorted(array, key=lambda x: x[0])

for i in array :
  print(i[0], i[1])
```
