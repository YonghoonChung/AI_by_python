# **1. 큐의 구조**


* 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 구조
* 줄을 서는 행위와 유사
* FIFO(First In, First Out)

### **1-1. 큐의 용어**

* Enque : 큐에 데이터를 넣는 기능
* Dequeue : 큐에 데이터를 꺼내는 기능

> 확인 : [Visualgo](https://visualgo.net/en/list
)

### **1-2. 파이썬  queue 라이브러리 활용해서 queue 자료구조 사용하기**

* Queue() : 가장 일반적인 큐 자료구조를 생성
* LifoQueue() : 나중에 입력된 데이터가 먼저 출력되는 구조의 큐 자료수조 생성(스택구조와 비슷)
* PriorityQueue(): 데이터마다 우선순위를 넣어서 우선순위를 높은 순으로 데이터를 출력
*프로그램을 작성할 때 프로그램에 따라 적합한 자료구조를 사용

```python
# Queue()로 큐 만들기
import queue #이미 있는 기능

data_queue = queue.Queue()
data_queue.put("Hello") #Enqueue

print(data_queue) #데이터의 주소를 알려준다
print(data_queue.qsize()) # 현재 요소가 몇개 있는지 리턴
print(data_queue.get()) #Dequeue #Hello가 날라감
print(data_queue.qsize())
print()
data_queue.put(10)
data_queue.put(True)
print('데이터 사이즈 : ' ,data_queue.qsize())
print(data_queue.get())
print('데이터 사이즈 : ' ,data_queue.qsize())

#queue 안에 있는 요소 출력
for i in range(data_queue.qsize()):
  item = data_queue.get()
  print(item)
```

```python
#LifoQueue()로 큐 만들기
import queue

data_queue = queue.LifoQueue()
data_queue.put('Hello')
data_queue.put(10)
data_queue.put(True)
print(data_queue.qsize())
print(data_queue.get())
```

```python
#PriorityQueue()로 큐 만들기
import queue

data_queue = queue.PriorityQueue()
data_queue.put((10, "김사과"))
data_queue.put((5, "반하나"))
data_queue.put((7, "오렌지"))
data_queue.put((8, "이메론"))
data_queue.put((2, "박체리"))
print(data_queue.qsize())
print(data_queue.get()) #키값이 적을 수록 우선순위가 높음
print(data_queue.get())
print(data_queue.qsize())
```

### **문제**
리스트 변수로 큐를 다루는 enqueue,dequeue 기능을 구현해보자

```python
queue_list = list()

def enqueue(data):
  queue_list.append(data)

def dequeue():
  return queue_list.pop(0)

enqueue(1)
enqueue(1)
enqueue(1)

print(queue_list)
```

```python
queue_list = list()

def enqueue(data):
  queue_list.append(data)

def dequeue():
  data = queue_list[0]
  del queue_list[0]
  return data

print(len(queue_list))

for index in range(10):
  enqueue(index)

print(queue_list)

print(dequeue())

print(queue_list)
```
