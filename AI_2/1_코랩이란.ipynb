# **2. 셀의 종류**

1. 코드 셀 : 코드 실행을 위한 셀
2. 텍스트 셀 : 주석을 달거나 설명, 문구등을 달기 위한 텍스트 전용 셀


### **2-1 셀의 삽입 방법**

1. 상단의 +코드, +텍스트 버튼을 눌러 삽입
2. 코드 셀 위에 삽입(컨트롤 + M, A)
3. 코드 셀 아래 삽입(ctrl + m , b)


### **2-2 셀의 변경 방법**


1. 코드 셀을 텍스트 셀로 변경 (ctrl + m, m)
2. 텍스트 셀을 코드 셀로 변경 (ctrl + m, y)

### **2-3 셀의 삭제**

1. 우측의 휴지통 아이콘을 클릭하여 삭제
2. ctrl + M , D


### **2-4 실행과 출력**

1. 컨트롤 + 엔터 : 해당 셀을 실행하고 커서는 해당 셀이 남음
2. 쉬프트 + 엔터 : 해당 셀을 실행하고 커서는 다음 셀로 넘어감
3. 알트 + 엔터 : 해당 셀을 실행하고 커서는 다음에 셀을 삽입하고 넘어감
