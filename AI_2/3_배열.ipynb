# **1.배열**

* 데이터를 나열하고 각 데이터를 인덱스에 대응하도록 구성한 자료구조
* 파이썬에서는 리스트 타입이 배열 기능을 제공함

### **1-1. 배열이 필요한 이유**

* 같은 종류의 데이터를 효율적으로 관리하기 위해 사용
* 같은 종류의 데이터를 순차적으로 저장
* 빠른 접근이 가능(인덱스 번호로 접근)

### **1-2. 배열의 단점**

* 데이터의 추가 / 삭제가 어려움
* 미리 최대 길이를 설정해야 함

```python
# 1차원 배열 : 리스트로 구현
arr_list = [1,2,3,4,5]
print(arr_list)
```

```python
arr_list = [[1,2,3],[4,5,6],[7,8,9]]
print(arr_list)
```

```python
print(arr_list[0])
print(arr_list[0][0])
```
