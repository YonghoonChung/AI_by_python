# **1. 수 정렬하기**
https://www.acmicpc.net/problem/2750


```python
# 데이터 개수 n을 입력받음
n = int(input())

# 리스트 선언
data = list()

# n개만큼 정수를 하나씩 입력받고 data에 append
for i in range(n):
  data.append(int(input()))

# 데이터가 모두 저장된 data를 오름차순으로 정렬
data.sort()

#정렬된 데이터를 한 줄씩 출력
for i in range(len(data)):
  print(data[i], end = '\n') 
  
```

# **2. 음계**
https://www.acmicpc.net/problem/2920

```python
arr =(input())
arrs = arr.split()
arr_list = list()
arr_list = [int(i) for i in arrs]

cnt = len(arr_list)
flag = 0
for i in range(cnt):
  for i in range(0,cnt-i-1):
    if(arr_list[i]<arr_list[i+1]):
      flag = 1
      if(arr_list[i+1]>arr_list[i+2]):
        flag = 3
    elif (arr_list[i]>arr_list[i+1]):
      flag = 2
      if(arr_list[i+1]<arr_list[i+2]):
        flag = 3
print(flag)
```

```python
num = list(map(int, input().split(' ')))

ascending = True
descending = True

for i in range(1, len(num)):
  if num[i] > num[i -1]:
    descending = False
  elif num[i] < num[i - 1]:
    ascending = False

if ascending:
  print('ascending')
elif descending:
  print('descending')
else:
  print('mixed')

```
