```python
import pandas as pd
```

```python
# 데이터프레임 받아오기
df = pd.read_csv('https://bit.ly/ds-house-price')
```

```python
#1. column의 이름을 재정의
df = df.rename(columns={'분양가격(㎡)': '분양가격'})
df
```

```python
# 2. 빈 값과 Data Type 확인
df.info()
```

```python
#3. 통계값 확인
df.describe()
# 분양가격은 숫자형이 아니란 것을 알수있음
```

```python
# 4. 분양가격 column을 int 타입으로 변환
df['분양가격'].astype(int)
```

```python
# 4-1. strip()을 활용하여 공백이 있는 데이터의 공백없애기
df['분양가격'] = df['분양가격'].str.strip();
```

```python
df.loc[df['분양가격'] == '  ']
```

```python
df
```

```python
df['분양가격'].astype(int)
```

```python
# 4-2. 공백이 있던 데이터를 모두 0으로 변경
df.loc[df['분양가격']=='', '분양가격'] = 0
```

```python
df.loc[df['분양가격'] == '']
```

```python
df['분양가격'].astype(int)
```

```python
# 4-3. NaN값을 fillna로 0을 채워줌
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)
```

```python
# 4-4, ,(콤마)가 들어간 데이터를 확인하고 제거
df.loc[df['분양가격'] == '6,657']
```

```python
df['분양가격'] = df['분양가격'].str.replace(',','')
```

```python
df.iloc[2125]
```

```python
df['분양가격'].astype(int)
```

```python
# 4-5. NaN값을 fillna로 0을 채워줌
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)
```

```python
# 4-6. -(하이픈) 없애준다
df['분양가격'] = df['분양가격'].str.replace('-','')
```

```python
df['분양가격'].astype(int)
```

```python
# 4-3. NaN값을 fillna로 0을 채워줌
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)
```

```python
df.loc[df['분양가격'] == '', '분양가격'] = 0
```

```python
df['분양가격'] = df['분양가격'].astype(int)
```

```python
df.info()
```

```python
# 5. 규모구분 column에 불필요한 '전용면접' 제거하기
df['규모구분'].value_counts()
```

```python
df['규모구분'] = df['규모구분'].str.replace('전용면적','')
```

```python
df['규모구분'].value_counts()
```

```python
# 6. 지역명 별로 평균 분양가격을 확인
df.groupby('지역명')['분양가격'].mean()
```

```python
# 7. 분양가격이 100보다 작은 행을 제거
df.loc[df['분양가격']<100]
```

```python
idx = df.loc[df['분양가격']<100].index
idx
```

```python
df = df.drop(idx, axis = 0)
df
```
